```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
data = pd.read_csv("train.csv")
df = data.drop("id", axis=1)
test_data = pd.read_csv("test.csv")
```

```python
columns = ["brand", "model", "model_year", "fuel_type", "engine", "transmission", "ext_col", "int_col", "accident", "clean_title"]
```

```python
encoder = OneHotEncoder(sparse_output=False)
transdf = encoder.fit_transform(df[columns])
```

```python
pd.DataFrame(transdf)
```

```python
dx = transdf
dy = df.price
```

```python
x_train, x_test, y_train, y_test = train_test_split(dx, dy, test_size=0.2)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```
